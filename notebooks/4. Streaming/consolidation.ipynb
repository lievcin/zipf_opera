```python
#database connection settings
import psycopg2

db_name = "traviato_development"
db_host = "localhost"
db_port = "5432"
db_user = "lievgarcia"
db_pwd = "traviato81"

conn = psycopg2.connect(database=db_name, user=db_user, password=db_pwd, host=db_host, port=db_port)
```

```python
import pandas as pd
import numpy as np

cursor = conn.cursor()

album_info_df = pd.read_csv('../../data/raw/streaming/dzr_albums.csv', index_col=0)
album_info_df = album_info_df.replace({r'\r\n': ''}, regex=True)

# album_info_df['composer'].str.strip(r'\\n')
# album_info_df['c'] = album_info_df.apply(lambda row: row['composer'].replace('\n', ''), index=1)

album_df = pd.read_csv('../../data/raw/streaming/dzr_isrc_song.csv', names=['album_id', 'song_isrc', 'song_id'])
song_df = pd.read_csv('../../data/raw/streaming/dzr_songs_isrc_listened.csv', names=['song_isrc', 'times_played'])

df = album_df.merge(album_info_df, left_on='album_id', right_on='deezer_id', how='outer')
df = df.merge(song_df, left_on='song_isrc', right_on='song_isrc', how='outer')

#Fill blank plays
df[['times_played']] = df[['times_played']].fillna(value=0)

# print(df.head())

o_df = df.groupby(['album_id'])
o_df = o_df.agg({'times_played': np.sum })
o_df = o_df.sort_values(by=['times_played'], ascending=False)
print(o_df.shape)
o_df.head(100)
```
