```python
from bs4 import BeautifulSoup
import requests
from datetime import datetime

headers = { 'User-Agent': 'Mozilla/5.0 (Windows NT 6.0; WOW64; rv:24.0) Gecko/20100101 Firefox/24.0' }

base_url = 'https://www.operaamerica.org/applications/NAWD/a-z.aspx'
```

```python
musical_works = []

r  = requests.get(base_url, headers=headers)
soup = BeautifulSoup(r.text, "html5lib")
results = soup.find(attrs={'id': 'topFullContentPlaceHolder_GridView1'}).find_all('tr')
results = results[1:]

for result in results:
    musical_work = []    
    columns = result.find_all('td')
    name    = columns[0].get_text()
    uri     = 'https://www.operaamerica.org/applications/NAWD/' + columns[0].find('a').get('href')
    
    r    = requests.get(uri, headers=headers)
    soup = BeautifulSoup(r.text, "html5lib")
    try:
        comp = soup.find(attrs={'id': 'blahblah'}).find('table').find('tr').find('strong').get_text()
        comp = comp.replace('       ', ' ')
    except:
        comp=''
#         print(uri)
    
    musical_work.append(name)
    musical_work.append(uri)  
    musical_work.append(comp)
    
    print(musical_work)
    musical_works.append(musical_work)
        
```

```python
#Saving results into dataframe and folder            
import pandas as pd                     
headers = ['name', 'uri', 'composer']
df = pd.DataFrame(musical_works, columns=headers)
df.to_csv('american_works.csv')
```
