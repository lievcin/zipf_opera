```python
from bs4 import BeautifulSoup
import requests
from datetime import datetime, date
from dateutil.relativedelta import relativedelta

headers = { 'User-Agent': 'Mozilla/5.0 (Windows NT 6.0; WOW64; rv:24.0) Gecko/20100101 Firefox/24.0' }

base_url = 'http://www.operascotland.org/tour/monthPanel/?type=list&mo='
```

```python
listings = []

for month in range(-748, 11):
    
    print(month)
        
    url = base_url + str(month)
    r  = requests.get(url, headers=headers)
    soup = BeautifulSoup(r.text, "html5lib")
    results = soup.find_all(attrs={'class': 'tour'})

    for result in results:
        
        if result.find(attrs={'class': 'title'}).get_text() != 'Sorry, we have no tours for this month.':
            
            #getting the name of the work            
            work_start = result.find(attrs={'class': 'title'}).get_text()[:-5]            
            work_url  = 'http://www.operascotland.org' + result.find(attrs={'class': 'title'}).find('a').get('href')
            work_r    = requests.get(work_url, headers=headers)
            work_soup = BeautifulSoup(work_r.text, "html5lib")
            work_page = work_soup.find(attrs={'class': 'content no tour'})
            work      = work_page.find('h3').find('a').get_text()
            if work == 'General':
                work = work_start            

            #getting info on the work
            work_url  = work_page.find('h3').find('a').get('href')
            work_url  = 'http://www.operascotland.org' + work_url
            work_r    = requests.get(work_url, headers=headers)
            work_soup = BeautifulSoup(work_r.text, "html5lib")            

            composer = ''
            librettist = ''            
            for item in list(zip(work_soup.find_all("dt"),work_soup.find_all("dd"))):
                role, person = item
                if role.get_text().strip() == 'Composer':
                    composer = person.get_text().strip()
                elif role.get_text().strip() == 'Librettist':
                    librettist = person.get_text().strip()               

            dates = result.find(attrs={'class': 'dates'}).find_all('span')
            for date in dates:
                listing = []
                theatre = date.get('title')

                #estimating date from the months
                year      = (datetime.today() + relativedelta(months=month)).date().year
                date      = date.text + ' ' + str(year)            
                date      = datetime.strptime(date, '%d %b %Y').date()                            

                listing.append(work)
                listing.append(composer)            
                listing.append(librettist)            
                listing.append(theatre)            
                listing.append(date)
                listings.append(listing)                    
            
#Saving results into dataframe and folder            
import pandas as pd                     
headers = ['work', 'composer', 'librettist', 'theatre', 'date']
df = pd.DataFrame(listings, columns=headers)
df.to_csv('../../data/processed/opera_scotland_listings.csv')            
```
