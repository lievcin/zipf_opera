```python
from bs4 import BeautifulSoup
import requests
from datetime import datetime, date
import locale
from dateutil.relativedelta import relativedelta

headers = { 'User-Agent': 'Mozilla/5.0 (Windows NT 6.0; WOW64; rv:24.0) Gecko/20100101 Firefox/24.0' }

base_url = 'https://archiv.wiener-staatsoper.at/search/page/'

locale.setlocale(locale.LC_ALL, 'de_DE')
```

```python
# listings = []
# counter = 0

for page in range(673, 812):
        
    url = base_url + str(page+1) + '?since=15.06.1700'
    r  = requests.get(url, headers=headers)
    soup = BeautifulSoup(r.text, "html5lib")
    results = soup.find_all(attrs={'class': 'simple-result-list-item'})

    for result in results:
        counter+=1
        listing = []
        work     = result.find('h2').get_text()
        work_url = 'https://archiv.wiener-staatsoper.at' + result.find('a').get('href')
        
        date     = result.find(attrs={'class': 'simple-result-list-item-details'}).get_text().strip()
        date     = datetime.strptime(date, '%A, %d. %B %Y').date()
        
        work_r    = requests.get(work_url, headers=headers)
        work_soup = BeautifulSoup(work_r.text, "html5lib")
        work_page = work_soup.find(attrs={'class': 'performance-date '})
        
        language = ''
        if 'Sprache' in work_page.find_all('p')[-1].get_text():
            language = work_page.find_all('p')[-1].get_text()
            listing.append(work)
            listing.append(date)
            listing.append(language)
            listing.append(work_url)
            listings.append(listing)
            
    print(page)

print(counter)            
print(len(listings))            
            
# #Saving results into dataframe and folder            
import pandas as pd                     
headers = ['work', 'date', 'language', 'work_url']
df = pd.DataFrame(listings, columns=headers)
df.to_csv('../../data/processed/vienna_opera_listings.csv')            
```
