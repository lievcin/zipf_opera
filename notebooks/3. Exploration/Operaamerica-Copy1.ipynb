```python
import pandas as pd                     

df = pd.read_csv('../../data/processed/operaamerica_listings.csv', index_col=0)
```

```python
df_c = df.groupby(['theatre'])['work'] \
         .count() \
         .reset_index(name='count') \
         .sort_values(['count'], ascending=False) \

# print(df_c[0:100])  
```

```python
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(5, 5), dpi= 100, facecolor='w', edgecolor='k')

plt.plot(list(df_c['count']))
plt.ylabel('Total performances')
plt.show()
```

```python
df_c = df.groupby(['work'])['theatre'] \
         .count() \
         .reset_index(name='count') \
         .sort_values(['count'], ascending=False)           

fig=plt.figure(figsize=(5, 5), dpi= 100, facecolor='w', edgecolor='k')
plt.plot(list(df_c['count']))
plt.ylabel('Total performances')
plt.show()
```
